### Outliers

```python
#find outlier of "Price" and "Annual Income"
def outliers_iqr(df):
    Q1 = np.percentile(df,25)
    Q3 = np.percentile(df,75)
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = [x for x in df if x < lower_bound or x > upper_bound]
    return outliers
```

```python
outlier_price = outliers_iqr(df['Price ($)'])
outlier_price
```

```python
outlier_income = outliers_iqr(df['Annual Income'])
outlier_income
```

## 2.2 Data Analysis

### The number of customers by gender

```python
plt.figure(figsize=(6,4))
sns.barplot(gender_count, x='Gender', y='count', palette="coolwarm")
plt.title("Number of customers by gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

- The number of male customerss who demand to use a car higher than female customers

#### The number of customers contributing to the transmission, color, engine

```python
categories = ['Transmission','Color','Engine']
titles = ["Transmission Type Distribution", "Car Color Distribution", "Engine Type Distribution"]
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 15))

# Loop through each category and create a pie chart
for i, category in enumerate(categories):
    count_data = df[category].value_counts().reset_index()
    count_data.columns = [category, 'Count']

    # Pie chart
    axes[i].pie(count_data['Count'], labels=count_data[category], autopct='%1.1f%%', colors=sns.color_palette("coolwarm", len(count_data)))
    axes[i].set_title(titles[i])

# Adjust layout for clarity
plt.tight_layout()
plt.show()
```

### The contribution of Car Price by Transmission, color, engine

```python
categories = ['Transmission', 'Color', 'Engine']
titles = ["Car Price Contribution by Transmission", "Car Price Contribution by Color", "Car Price Contribution by Engine"]

# Create subplots for pie charts
fig, axes = plt.subplots(3, 1, figsize=(8, 15))

# Loop through each category and create a pie chart
for i, category in enumerate(categories):
    grouped_data = df.groupby(category)[ 'Price ($)'].sum()  # Sum Car Price per category

    axes[i].pie(
        grouped_data, 
        labels=grouped_data.index, 
        autopct='%1.1f%%', 
        colors=sns.color_palette("coolwarm", len(grouped_data)), 
        startangle=140
    )
    axes[i].set_title(titles[i])

plt.tight_layout()
plt.show()
```
