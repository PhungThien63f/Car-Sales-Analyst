### Outliers

```python
#find outlier of "Price" and "Annual Income"
def outliers_iqr(df):
    Q1 = np.percentile(df,25)
    Q3 = np.percentile(df,75)
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = [x for x in df if x < lower_bound or x > upper_bound]
    return outliers
```

```python
outlier_price = outliers_iqr(df['Price ($)'])
outlier_price
```

```python
outlier_income = outliers_iqr(df['Annual Income'])
outlier_income
```

# 3.Modelling & Evaluation

## Feature Engineering

```python
#choosing feature
df = df.drop(labels=['Car_id','Gender','Date','Customer Name','Annual Income','Dealer_No ','Phone','Dealer_Name'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

def apply_feature_engineering(df):
    le = LabelEncoder()
    for column in df.columns:
        if df[column].dtype == "object":
            df[column] = df[column].astype(str)
            df[column] = le.fit_transform(df[column])
    return df

df = apply_feature_engineering(df)
```

```python
df
```

## Model

### Train-test split

```python
X = df.drop(['Price ($)'],axis=1)
y = df['Price ($)']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test ,y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
from sklearn.metrics import mean_absolute_percentage_error
```

### Scaler data

```python
from sklearn.preprocessing import StandardScaler

# scaler data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train.astype(np.float64))
X_test = scaler.transform(X_test.astype(np.float64))
```

### Linear Regression 

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

# Predict
y_pred = linear_reg.predict(X_test)

# Evaluation
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
explained_var = explained_variance_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100
accuracy = 100 - mape  

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R² Score): {r2:.4f}")
print(f"Explained Variance Score: {explained_var:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
print(f"Model Accuracy: {accuracy:.2f}%")
```

### Ramdom Forest  

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)

# Predict
y_pred = rf_model.predict(X_test)

# Evaluation
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
explained_var = explained_variance_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100
accuracy = 100 - mape  

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R² Score): {r2:.4f}")
print(f"Explained Variance Score: {explained_var:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
print(f"Model Accuracy: {accuracy:.2f}%")
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor


# XGBoost Model
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)

# Predict
y_pred = dt_model.predict(X_test)

# Evalution
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
explained_var = explained_variance_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100
accuracy = 100 - mape 


print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R² Score): {r2:.4f}")
print(f"Explained Variance Score: {explained_var:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
print(f"Model Accuracy: {accuracy:.2f}%")
```

### XGBoost

```python
from xgboost import XGBRegressor 

# scaler data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train.astype(np.float64))
X_test = scaler.transform(X_test.astype(np.float64))

# XGBoost Model
xgboost = XGBRegressor()
xgboost.fit(X_train, y_train)

# Predict
y_pred = xgboost.predict(X_test)

# Evalution
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
explained_var = explained_variance_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100
accuracy = 100 - mape 


print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R² Score): {r2:.4f}")
print(f"Explained Variance Score: {explained_var:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
print(f"Model Accuracy: {accuracy:.2f}%")
```
